```python
import pandas as pd
```

```python
df = pd.read_excel("NVDA_Q.xls")
df.dropna(how='all', inplace=True)
```

```python
df
```

```python
index_PL = df.loc[ df["Data provided by SimFin"] == 'Profit & Loss statement'].index[0]

print(index_PL)
```

```python
index_CF = int(df.loc[df['Data provided by SimFin']=='Cash Flow statement'].index[0])
index_BS = int(df.loc[df['Data provided by SimFin']=='Balance Sheet'].index[0])

print(index_CF)
print(index_BS)
```

```python
df_PL = df.iloc[index_PL:index_BS-2, 1:]
print(df_PL)
```

```python
df_PL.columns = df_PL.iloc[0]
```

```python
df_PL
```

```python
df_PL.set_index("in million USD",inplace=True)
```

```python
df_PL
```

```python
df_PL.fillna(0, inplace=True)
df_PL
```

```python
# these are all same
#df_PL[0:] = df_PL = df[:]
df_PL=df_PL[1:]
df_PL
```

```python
df_BS = df.iloc[index_BS-1:index_CF-3, 1:]
df_BS
```

```python
df_BS.columns = df_BS.iloc[0]
df_BS
```

```python
df_BS.set_index("in million USD", inplace=True)
```

```python
df_BS
```

```python
df_BS.fillna(0, inplace=True)
df_BS
```

```python
df_BS = df_BS[1:]
```

```python
df_BS
```

```python
df_CF = df.iloc[index_CF-2:, 1:]
df_CF
```

```python
df_CF.columns = df_CF.iloc[0]
df_CF
```

```python
df_CF = df_CF[1:]
df_CF
```

```python
df_CF.set_index("in million USD", inplace=True)
```

```python
df_CF
```

```python
df_CF.fillna(0, inplace=True)
df_CF
```

```python
df_PL.to_pickle("NVDA_Q_IS.pkl")
df_BS.to_pickle("NVDA_Q_BS.pkl")
df_CF.to_pickle("NVDA_Q_CF.pkl")
```
